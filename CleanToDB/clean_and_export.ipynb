### CLEANING DATA

```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
import os
```

```python
institutions_csv = pd.read_csv(os.path.join("rawdata", "institutions.csv"), index_col="UnitID")
```

```python
# Create Actual SQL Lite DB
conn = sqlite3.connect(os.path.join('..','api','static','db','studentloans.sqlite'))
```

```python
inst_df = institutions_csv
inst_df.reset_index
inst_df.head()
```

```python
inst_df.dtypes
```

```python
# Create an in memory sql lite engine
engine = create_engine('sqlite:///../api/static/db/studentloans.sqlite', echo=False)
```

```python
# Create an institutions table and put it the csv in in
inst_df.to_sql('institutions', con=engine)
```

```python
engine.execute("SELECT * FROM institutions").fetchall()
```
