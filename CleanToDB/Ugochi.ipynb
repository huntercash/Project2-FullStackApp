```python
import pandas as pd
import os
```

```python
institutions_csv = pd.read_csv(os.path.join("rawdata", "institutions.csv"), index_col="UnitID")
institutions_csv.head()
```

```python
financial_aid= pd.read_csv(os.path.join("rawdata", "financial-aid.csv"), index_col="UnitID")
financial_aid.head()
```

```python
financial_aid= financial_aid.fillna(0)
financial_aid.head()
```

```python
joined_data= institutions_csv.join(financial_aid)
joined_data.head()
```

```python
joined_data= joined_data.drop(columns=["Institution Name", "Institution name alias (HD2016)", 
                                                 "Street address or post office box (HD2016)", "City location of institution (HD2016)",
                                                "State abbreviation (HD2016)"])
joined_data.head()
```

```python
joined_data.to_csv("rawdata/institution_loans.csv")
```

```python
groupby_state= joined_data.groupby(["state"]).sum()
groupby_state.head()
```

```python
loans_by_state= groupby_state[["total_seeking_aid","number_awarded_loans", "Amount_loans_awarded"]]
loans_by_state.head()
```

```python
statelatlong= pd.read_csv(os.path.join("rawdata", "statelatlong.csv"), index_col="State")
statelatlong.head()
```

```python
joined_data2= statelatlong.join(loans_by_state)
joined_data2
```
