```python
import pandas as pd
```

```python
portfolio_summ =  pd.read_excel('Data/PortfolioSummary.xls')
```

```python
portfolio_summ.columns
```

```python
new_header = ["Year","Quarter",'Direct Loans','DL Recipients',
'FFEL','FFEl Recipients',
'Perkins','Perkins Recipients',
'Total loans','Total Recipients']
    
```

```python
portfolio_summ = portfolio_summ[5:]
portfolio_summ.columns = new_header
portfolio_summ
```

```python
portfolio_summ = portfolio_summ[:-5]
```

```python
portfolio_summ = portfolio_summ.fillna(method='ffill')
```

```python
portfolio_summ = portfolio_summ.drop(columns=['Quarter'])
```

```python
portfolio_summ = portfolio_summ.reset_index(drop=True)
portfolio_summ
```

```python
yearly_portfolio_summ = portfolio_summ.groupby(['Year']).sum()
yearly_portfolio_summ
```

```python
yearly_portfolio_summ.dtypes
```

```python
yearly_portfolio_summ.to_csv('clean_data/yearly_portfolio_summary.csv', index=True)
```
